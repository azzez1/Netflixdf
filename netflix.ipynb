```python
import pandas as pd
import numpy as np
from pandas import DataFrame, Series
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
flixdf=pd.read_csv('Netflix_titles.csv')
flixdf.head()
```

```python
#selecting a specific term of the column from the DF
# Getting specific Information from the coloums and making another df to save them.


ind=flixdf.query('country=="India"')
ind
```

```python
### creating a new column while assingning a number from given columns only

flixdf['Month']=flixdf['date_added'].str[:3]
flixdf.tail()
```

```python
flixdf['listed_in'].value_counts()
```

```python
# Using group by to select specific columns AND data.   LEARN IMPORTANT THINGS

ratings= flixdf.groupby('rating')['release_year'].sum().sort_values(ascending=True).to_frame()
ratings
```

```python
fig=plt.figure(figsize=(10,5),dpi=200)
axes=fig.add_axes([0.1,0.1,0.8,0.8])
axes.bar(ratings['rating'],ratings["release_year"],color='blue')
plt.show()
```
